### 2.2 Load the data file

The following commands load the data file "Supporting information-3.xlsx" to our programme. ***The data file should be put in the same directory as this notebook*** for the code to run. There are 132 rows and 39 columns in the data, excluding the header and the index. The last column records the actual wavenumbers of characteristic peaks of the carbonyl groups in infrared spectral. The other columns record different functional groups attached to the carbonyl group, where 0 stands for "not exist" and 1 stands for "exist". 

In the header, you will notice that some functional group names are typed in formula form using $\LaTeX$. This is for proper display of functional group names in the figures. Here are some examples on $\LaTeX$ basics:

(1) `CH2CH3` -> CH2CH3
<br>
Not preffered: Is not in formula form, therefore no subscription.

(2) `$CH_{2}CH_{3}$` -> $CH_{2}CH_{3}$
<br>
Not preffered: Is in formula form, but "CH" is recongnised as variables rather than plain texts.

(3) `$\text{CH}_{2}\text{CH}_{3}$` -> $\text{CH}_{2}\text{CH}_{3}$ 
<br>
Preffered: Is in formula form, and "CH" is recongnised as plain texts.

Additionally, writng plain text part out of `$$` is also acceptable in the DataFrame. For example, both `$\text{CH}_{2}\text{Br}$` and `$\text{CH}_{2}$Br` can be used to describe $\text{CH}_{2}\text{Br}$.

Moreover, please note that the column names used by `pandas` may not be the same as displayed in the DataFrame because of the use of $\LaTeX$. You should always use the column names given by `data.columns.tolist()` when indexing.

```python
data = pd.read_excel('Supporting information-3.xlsx', header = 0, sheet_name = 0) # Read the data in the first spreadsheet and set the first row as header.
data = data.dropna().reset_index(drop=True) # Drop rows with NaN values and reset the order of the index.
print('Columns in the dataset:\n', data.columns) # "\n" starts a new line.
print('Columns used by pandas:\n', data.columns.tolist())
print('Index:\n', data.index)
data
```

### 2.3 Data pre-processing

After reading the data file, we need to pre-process the data. Firstly, we define X and y. In this task, X is composed of the functional group while y is their wavenumbers of characteristic peaks of the carbonyl groups. Therefore, X is a $132 \times 38$ matrix and y is a $132 \times 1$ vector. In machine learning, we usually call each input element a "feature". Thus, we have 38 features here. Next, we split the full dataset into a training set and a hold-out test set. In subsequent steps, the training set will be used for model development while the test set will be used for model evaluation. The test set cannot be used for model development in any case.

Finally, we use standard scaler to scale X by subtract each feature's mean value and divided by its standard deviation. The mean and standard deviation are calculated using training X only but are applied to both training and test X. Scaling is one of the most important pre-processing step for many machine learning tasks and the requirement of many machine learning models. This process makes all input features to follow $\mathcal{N}(0,1)$ distribution. Therefore, features with notable difference in the order of magnititude are adjusted to similar scopes by scaling, which would significantly reduce the influence brought by different feature values. However, it should be noted that the orginal data is already in the same scope (0-1 values) in this project. As a result, scaling is not compulsary here. Nevertheless, we keep this step for the purpose of teaching a more general machine learning workflow.

```python
# Define X and y.
X = data.loc[:, '$\\text{CH}_{2}\\text{CH}_{3}$':'H'] # Taking out every line and column "$\text{CH}_{2}\text{CH}_{3}$" to column "H" (both sides are included).
y = data['Actual']

# Train-test split.
## 1958 is the founding year of USTC.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1958)

# Scale X.
S = StandardScaler().fit(X_train.values)
X_train = S.transform(X_train.values)
X_test = S.transform(X_test.values)
```

### 2.4 Model training and evaluation

We will cover the training and evaluation of three models in this section, namely linear regression (LR), support vector regression (SVR), and random forest (RF).

We start with the simplest case - linear regression. The steps are:

(1) Define and fit the model: Define the LR model using `LinearRegression()` we imported in Section 2.1 and use training data to fit the model. `m1` is the LR model fitted on training data. You can refer to https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html for more details.
<br>
(2) Find the predictions made by the fitted model `m1` and the test input `X_test`. The predictions are defined as `m1_pred`. It has exactly the same dimensions as `y_test`.
<br>
(3) Evaluate the test results using mean absolute error (MAE) and $R^2$ score.
<br>
(4) Furthermore, we should also have a look at the model's performance on the training set. It is a good thing if the model performs similarly on the training and test datasets. If the model performs badly on both sets, there is a sign of underfitting, which suggests that the model is not good enough for the current task. If the model performs much better on the training set, there is a sign of overfitting. In other words, the model memorizes the training data rather than learns patterns from them, and perform badly on unseen datasets. If the model performs much better on the test set, you might have made some mistakes or ran into a coincidence.

```python
# Define and fit the model.
m1 = LinearRegression().fit(X_train, y_train)

# Find predictions.
m1_pred = m1.predict(X_test)

# Evaluate the results on test data.
print('Test MAE:', mean_absolute_error(y_test, m1_pred))
print('Test R^2 score:', r2_score(y_test, m1_pred))

# Evaluate the results on training data.
m1_pred_training = m1.predict(X_train)
print('Training MAE:', mean_absolute_error(y_train, m1_pred_training))
print('Training R^2 score:', r2_score(y_train, m1_pred_training))
```

Next, we visualize the results using `Matplotlib`. In the following figure, the X-axis represents actual values while the Y-axis represents predicted values. The reference line represents Y=X. Therefore, a point on the line stands for a perfect prediction. In general, points closer to the line reflects better results.

```python
# Visualization
plt.figure(figsize=(6,5))
plt.scatter(y_train, m1_pred_training, marker='o', s=16, color='dodgerblue', alpha=0.4, edgecolors='none', label='Training set', zorder=1)
plt.scatter(y_test, m1_pred, marker='^', s=40, color='darkgreen', label='Test set', zorder=2)
plt.plot([1600, 1850], [1600, 1850], 'k-', alpha=0.75, label='Reference line', zorder=0)
plt.xlabel('Actual values', fontsize=15)
plt.ylabel('Predicted values', fontsize=15)
plt.title('LR results', fontsize=20)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('lr_results.png')
plt.show()

# Save to file
lr_result = pd.DataFrame({'Actual': np.concatenate((y_train, y_test)), 'Pred': np.concatenate((m1_pred_training, m1_pred))})
lr_result.to_excel('lr_results.xlsx', index=False)
```

Next, let's explore more complicated machine learning models. The first one is support vector regression (SVR). The official documentation of `scikit-learn`'s SVR is available on https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html . From the official documentation of SVR, we can find that there are many configurable parameters for the SVR model. For example, `kernel` specifies the kernel type to be used in the model. These parameters are called "hyperparameters" in machine learning context. This is to distinguish model configurations from model parameters (for example, weights in LR) and input parameters (usually called features in machine learning).

To address the influence of hyperparameters on model performance, we apply grid search with cross-validation (GSCV) for hyperparameter tunning. The purpose of hyperparameter tunning is to find the group of hyperparameters that best fits the training data. To understand GSCV, we have to understand both grid search with cross-validation.

Grid search: Consider there are $N$ hyperparameters to be tunned, we can regard each hyperparameter combination as a point on the $N$-dimensional plane. Therefore, all hyperparameter combinations form a grid in that plane. Then, we calculate model performance on each point on the grid and find the point where our model has the best performance. The respective hyperparameter combination will be used for subsequent modelling.

Cross-validation is the technique to determine which parameter combination gives the best model performance. To conduct cross-validation, we first divide the full training set into smaller parts with equal size. If we divide the full training set into K parts, we call the cross-validation process a K-fold validation. In each fold, we train the model with a hyperparameter combination using the  (K-1)/K of the joint set and evaluate its performance on the remaining 1/K. Finally, evaluation results from all folds are combined as the final evaluation result of this group of hyperparameters.

For more details on GSCV, please visit the official documentation on https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html . Please note that `n_jobs`, which determines the number of concurrent processes or threads to be used, is set to be -1 in this notebook, meaning that all available computational resources are used. However, this setting may fail on some computers. In this case, it is recommanded that you alter -1 to another integer according to you CPU's capacity specifications. Another point to address is that, when running GSCV, you may see a "RuntimeWarning" in the output. This is usually caused by improper hyperparameter combinations during the search and can be ignored.

```python
# Define the hyperparameter groups to search
param_grid_2 = {
  'C': np.linspace(0.1, 2, 20), # [0.1, 0.2,..., 2.0]
  'epsilon': np.linspace(0.2, 10, 50)  # [0.2, 0.4,..., 10.0]
}

# Define the model and the hyperparameter you do not wish to tune. SVR with a linear kernel can be regarded as a mroe advanced version of LR.
svr = SVR(kernel='linear')

# GSCV
grid_search_2 = GridSearchCV(estimator = svr, param_grid = param_grid_2, scoring='neg_mean_squared_error', cv = 5, n_jobs = -1)
grid_search_2.fit(X_train, y_train)

# GSCV results
## The first line is to force a clean output, or 1.1 may appear as 1.0999999999999999.
best_params = {k: round(v, 2) if isinstance(v, float) else v for k, v in grid_search_2.best_params_.items()}
print("The best hyperparameter combination given by GSCV is:", best_params)
```

Model fitting, evaluation, and visualization follow the same process as LR.

```python
# Redefine the model using the best hyperparameters and fit the model.
m2 = SVR(kernel='linear', **grid_search_2.best_params_).fit(X_train, y_train)

# Find predictions.
m2_pred = m2.predict(X_test)

# Evaluate the results on test data.
print('Test MAE:', mean_absolute_error(y_test, m2_pred))
print('Test R^2 score:', r2_score(y_test, m2_pred))

# Evaluate the results on training data.
m2_pred_training = m2.predict(X_train)
print('Training MAE:', mean_absolute_error(y_train, m2_pred_training))
print('Training R^2 score:', r2_score(y_train, m2_pred_training))

# Visualization
plt.figure(figsize=(6,5))
plt.scatter(y_train, m2_pred_training, marker='o', s=16, color='dodgerblue', alpha=0.4, edgecolors='none', label='Training set', zorder=1)
plt.scatter(y_test, m2_pred, marker='^', s=40, color='darkgreen', label='Test set', zorder=2)
plt.plot([1600, 1850], [1600, 1850], 'k-', alpha=0.75, label='Reference line', zorder=0)
plt.xlabel('Actual values', fontsize=15)
plt.ylabel('Predicted values', fontsize=15)
plt.title('SVR results', fontsize=20)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('SVR_results.png')
plt.show()

# Save to file
svr_result = pd.DataFrame({'Actual': np.concatenate((y_train, y_test)), 'Pred': np.concatenate((m2_pred_training, m2_pred))})
svr_result.to_excel('svr_results.xlsx', index=False)
```

Then, we try the random forest regression model (RF). The official documention can be found on https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html . 

The modelling with RF follows the same steps as SVR, so we skip detailed explainations here. The code is given below. Please bear in mind that RF runs significantly slower than LR and SVR.

```python
# Define the hyperparameter groups to search
param_grid_3 = {
  'max_depth': list(np.linspace(2,30,15, dtype=int)),
  'max_features': list(np.linspace(2,30,15, dtype=int)),
  'n_estimators': list(np.linspace(2,30,15, dtype=int))
}

# Define the model and the hyperparameter you do not wish to tune.
rf = RandomForestRegressor(bootstrap=True, random_state=1958)

# GSCV
grid_search_3 = GridSearchCV(estimator = rf, param_grid = param_grid_3, scoring='neg_mean_squared_error', cv = 5, n_jobs = -1)
grid_search_3.fit(X_train, y_train)

# GSCV results
best_params = {k: round(v, 2) if isinstance(v, float) else v for k, v in grid_search_3.best_params_.items()}
print("The best hyperparameter combination given by GSCV is:", best_params)

# Redefine the model using the best hyperparameters and fit the model.
m3 = RandomForestRegressor(bootstrap=True, random_state=1958, **grid_search_3.best_params_).fit(X_train, y_train)

# Find predictions.
m3_pred = m3.predict(X_test)

# Evaluate the results on test data.
print('Test MAE:', mean_absolute_error(y_test, m3_pred))
print('Test R^2 score:', r2_score(y_test, m3_pred))

# Evaluate the results on training data.
m3_pred_training = m3.predict(X_train)
print('Training MAE:', mean_absolute_error(y_train, m3_pred_training))
print('Training R^2 score:', r2_score(y_train, m3_pred_training))

# Visualization
plt.figure(figsize=(6,5))
plt.scatter(y_train, m3_pred_training, marker='o', s=16, color='dodgerblue', alpha=0.4, edgecolors='none', label='Training set', zorder=1)
plt.scatter(y_test, m3_pred, marker='^', s=40, color='darkgreen', label='Test set', zorder=2)
plt.plot([1600, 1850], [1600, 1850], 'k-', alpha=0.75, label='Reference line', zorder=0)
plt.xlabel('Actual values', fontsize=15)
plt.ylabel('Predicted values', fontsize=15)
plt.title('RFR results', fontsize=20)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('rfr_results.png')
plt.show()

# Save to file
rfr_result = pd.DataFrame({'Actual': np.concatenate((y_train, y_test)), 'Pred': np.concatenate((m3_pred_training, m3_pred))})
rfr_result.to_excel('rfr_results.xlsx', index=False)
```

### 2.6 Explain models using SHAP

In the last part of this project, we will quantify the shift patterns of different functional groups to carbonyl peak shifts in IR spectra using SHapley Additive exPlanations (SHAP). SHAP was proposed by Lundberg and Lee and the orginal paper can be accessed from https://dl.acm.org/doi/10.5555/3295222.3295230 . There is also a great book regarding explainable machine learning by Molnar, which can be accessed from https://christophm.github.io/interpretable-ml-book/ . Chapters 17 and 18 contain introduction to SHAP. In short, SHAP decomposes the final model output to the sum of a base output and each feature's contribution towards the final output. We will discuss this further with an example shortly.

As for now, let's focuse on how to acquire SHAP results. Section 2.4 has demonstrated that SVR is the best model in this project. Therfore, we use SVR with the best parameters to build the final model. The final model is built on the full dataset using the scaling parameters calculated from the training set. Then, we use `KernelExplainer`, which is an explainer that can be applied to any model, to derive SHAP explanations. The resulting `shap_explanation` is composed of three NumPy arrays (You can regard them as matrices). `shap_explanation.values` contains the SHAP values, `shap_explanation.base_values` contains base outputs of the model, and `shap_explanation.data` contains the input data. All matrices has 132 rows, each corresponding to one row in X.

To supplement, the reason to use scaling parameters calculated from the training set is that the best SVR model is developed from this setting. The reason to use the full dataset is that we would like to quantify the influence of EVERY functional groups we have. Using merely training or test set may lead to a decrease in the available number of functional groups. If you are working with a project with a large amount of data in your future research, you can fit the model and the SHAP explainer using the training data only.

```python
# Fit the best model using the full dataset
X_scaled = S.transform(X.values)
model = SVR(kernel='linear', **grid_search_2.best_params_).fit(X_scaled, y)

# Calculate SHAP explanation
explainer = shap.KernelExplainer(model.predict, X_scaled, feature_names=X.columns)
shap_explanation = explainer(X_scaled, l1_reg='num_features(38)') # num_features(38) means considering all 38 features. The default number is 10.
# You may see a warning message "Could not render content for 'application/vnd.jupyter.widget-view+json'" below before running the code. It can be safely ignored and will vanish after running the code on yourself.
```

Next, we are going to explain the SHAP results. Firstly, we investigate waterfall plot of the first row. In the original data set, the first row represent 2-butanone, whose wavenumber is 1718 . The figure indicates that the final model output is 1715 and the base output is 1711.788. The existance of $\text{CH}_2\text{CH}_3$ makes the final output 13.73 larger while the $\text{CH}_3\_\text{r}$ pushes the final output 9.51 towards the neagative direction. The influence of other functional groups on the final output are also displayed in the figure.

```python
shap_explanation.data = X.values # This is to subsitute scaled X with orginal X for better display in the figure.
shap.plots.waterfall(shap_explanation[0], max_display=99)
```

SHAP's plotting procedure cannot be precisely controlled by users and it may cause problems during saving. Therefore, we re-plot the previous figure for saving.

Secondly, we investigate the summary plot. The summary plot is the summary of all features' SHAP values regarding all data rows explained. We would get a summary plot that seems to have two "lines" in each row. Red represent existance and bule represents non-existance. Actually, each "line" is composed of repetitive points with the same value and the total number of points in each row of the summary plot equals 132. The reason why we have "lines" is that our SVR model uses a linear kernel with 0-1 data. In other cases, a more scattered summary plot is expected. You can refer to this official documentation as an example: https://shap.readthedocs.io/en/latest/example_notebooks/tabular_examples/model_agnostic/Diabetes%20regression.html .

```python
shap_values = shap_explanation.values
shap.summary_plot(shap_values, X_scaled, feature_names=X.columns, max_display=99)
```

Next, we are going to visualize the average difference in the final output made by a functional group changing from non-existant to existant. That is, the difference (not the absolute difference!) between red and blue points in each row of the summary plot. The first step is to calculate a DataFrame from the SHAP results. This DataFrame will be used for figure plotting.

```python
print('Shape of shap_values:', shap_values.shape)
n_cases = shap_values.shape[0] # Number of rows
n_features = shap_values.shape[1] # Number of input features
differences = []
for i in range(n_features):
    exist_shap = []
    nonexist_shap = []
    for j in range(n_cases):
        if data.iloc[j, i] == 1:
            exist_shap.append(shap_values[j, i])
        else:
            nonexist_shap.append(shap_values[j, i])
    differences.append(np.mean(exist_shap)-np.mean(nonexist_shap))
differences_df = pd.DataFrame({'Group': X.columns, 'Difference': differences})
differences_df = differences_df.sort_values(by='Difference').reset_index(drop=True)
differences_df
```
